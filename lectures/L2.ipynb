```python
import requests
from bs4 import BeautifulSoup
#Biden speech
URL = "https://www.whitehouse.gov/briefing-room/speeches-remarks/2021/01/20/inaugural-address-by-president-joseph-r-biden-jr/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find(id="content")
biden_speech = list(results.find_all("p"))
biden_speech = [str(x) for x in biden_speech]
```

```python
biden_speech[:5]
```

```python
#format biden_speech
content = "".join(biden_speech)
print(content)
```

```python
#clean the text 1
cleaned_speech = re.sub("<.*?>", "", content)
cleaned_speech
```

```python
#clean the text 2
cleaned_speech = re.sub("<.*?>", "", content)
cleaned_speech = re.sub(".*THE PRESIDENT:","",content)
cleaned_speech = re.sub("END.*","",content)
```

```python

print(len(re.findall(r"\bAmerica\b",cleaned_speech)))
print(len(re.findall(r"\bAmerica",cleaned_speech)))
print(len(re.findall(r"\bwe\b",cleaned_speech.lower())))
print(len(re.findall(r"\bi\b",cleaned_speech.lower())))
print(len(re.findall(r"i",cleaned_speech.lower())))
```

```python
print(len(re.findall(r"\bAmerica\b",cleaned_speech)))
print(len(re.findall(r"\bAmerica",cleaned_speech)))
print(len(re.findall(r"\b[wW]e\b",cleaned_speech)))
print(len(re.findall(r"\bi\b",cleaned_speech.lower())))
print(len(re.findall(r"i",cleaned_speech.lower())))
```
